```python
# Q7
j = input("Please enter your text: ")
def encode(j):

    vowel = ['a', 'e', 'i', 'o', 'u','A','E','I','O','U']
    j = j.split(" ")
    x = []

    for i in range(len(j)):
        if j[i][0] in vowel:
            new = j[i][0] + j[i][-1]
            x.append(new)
        else:
            for v in vowel:
                j[i] = j[i].replace(v, "")
            x.append(j[i])

    print('Your encoded message is:',' '.join(x))

encode(j)
```

```python



def categorize_bmi(bmi):
    if bmi < 19.5:
        return 'Underweight'
    elif bmi <= 25:
        return 'Normal weight'
    elif bmi <= 30:
        return 'Overweight'
    else:
        return 'Obese'

# Apply the function to create a new column for BMI categories
df['BMI Category'] = df['Body mass index'].apply(categorize_bmi)
df
```
