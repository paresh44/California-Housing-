6. Write a programme to  construct (create a user defined function) to calculate the median value of the data set wherever required.



```python
def median(data):
  data1 = sorted(data)
  if len(data1)%2==0:
    x = data1[len(data1)//2]
    y = data1[(len(data1)//2)-1]
    median = (x+y)/2
  else:
    median = data1[len(data1)//2]
  print(median)

```

```python
s=pd.Series([1,2,3,4,10,11,12])
median(s)
# s.median()
```
